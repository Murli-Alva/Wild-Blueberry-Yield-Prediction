```python
import pandas as pd                         # importing pandas library
from datetime import datetime
from werkzeug.utils import secure_filename
import os

df = pd.read_excel('student_profile.xlsx', index_col=0)  # reading data file
df["Name"] = [i.strip() for i in df["Name"].values]
df["Interested_Domian"] = [i.strip() for i in df["Interested_Domian"].values]
df["Current Education"] = [i.strip() for i in df["Current Education"].values]
df
```

```python
unq = df.groupby("Interested_Domian")['Name'].count() # counting the number of interested domains
unq.head(10)
```

```python
z = input('Enter size of group: ')    # specifying the size of the group
x = input('Enter domain: ')           # specifying the domain of the group
df1 = df[df["Interested_Domian"]== x].sample(n = int(z))

df2 = df[df["Interested_Domian"]== x] # checking if the domain has any student left to be grouped
if len(df)>0:
    for i in df1.index.values:
        df2.drop(i, inplace=True)
```

```python
if not os.path.exists("groups"): # making the directory for storing the group file
    os.makedirs("groups")
df1.to_excel("groups/"+secure_filename("Group_{}".format(datetime.now()))+".xlsx") # File of group students
df2.to_excel("groups/"+secure_filename("Remaining_people_Group_{}".format(datetime.now()))+".xlsx") # File of students remaing to be grouped
df1
```
